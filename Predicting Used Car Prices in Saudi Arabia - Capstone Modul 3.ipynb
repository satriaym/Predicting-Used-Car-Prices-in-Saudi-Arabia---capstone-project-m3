```python
# Creating the heatmap
plt.figure(figsize=(8, 6))  # Set the size of the heatmap
sns.heatmap(cars.corr(numeric_only=True), annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Display the plot
plt.title('Correlation Between Numeric Columns With Price')
plt.show()
```

```python
def correlation_ratio(categories, values):
    values = np.array(values)
    categories = np.array(categories)
    unique_categories = np.unique(categories)
    category_means = [values[categories == cat].mean() for cat in unique_categories]
    overall_mean = values.mean()
    ss_between = sum(
        len(values[categories == cat]) * (mean - overall_mean) ** 2
        for cat, mean in zip(unique_categories, category_means)
    )
    ss_total = sum((values - overall_mean) ** 2)
    return np.sqrt(ss_between / ss_total)

# Mengambil kolom kategorikal
objects = cars.select_dtypes(include=['object']).columns

# Menyimpan nilai korelasi
features = []

# Menghitung correlation ratio untuk setiap kolom kategorikal dengan Price
for col in objects:
    corr = round(correlation_ratio(cars[col], cars['Price']), 2)
    features.append(corr)

# Membuat DataFrame hasil
df_ratio = pd.DataFrame(
    data=[features], 
    columns=list(objects), 
    index=['Price']
)

# Menampilkan heatmap
plt.figure(figsize=(16, 2))
sns.heatmap(df_ratio, annot=True, cbar=False, cmap='coolwarm')
plt.title('Correlation Ratio of Categorical Features with Price')
plt.show()
```

### 5.1.3 Hyperparameter Tuning on XGBoost Regressor

Performing hyperparameter tuning on the XGBoost Regressor for several reasons:

1. **Improve Model Accuracy**:
   - Adjusting hyperparameters like `learning_rate`, `max_depth`, and `n_estimators` helps the model make more accurate predictions.

2. **Reduce Overfitting**:
   - By tuning parameters like `max_depth` and `min_child_weight`, I can reduce overfitting and improve the model's ability to generalize.

3. **Enhance Training Efficiency**:
   - Parameters such as `subsample` and `colsample_bytree` speed up the training process without compromising performance.

4. **Improve Generalization Ability**:
   - Regularization parameters like `lambda` and `alpha` help prevent overfitting and improve the model’s ability to generalize well on unseen data.

Hyperparameter tuning ensures that XGBoost delivers **the best results, optimizing accuracy, efficiency, and generalization.**

```python
# Tree depth (max_depth)
max_depth = list(np.arange(3, 11))  # Values from 3 to 10 for depth of the tree

# Learning rate (learning_rate)
learning_rate = list(np.arange(1, 31)/100)  # Values from 0.01 to 0.3 for learning rate

# Number of trees (n_estimators)
n_estimators = list(np.arange(100, 1001))  # Values from 100 to 1000 for number of estimators

# Fraction of rows per tree (% from total training set) (subsample)
subsample = list(np.arange(2, 10)/10)  # Values from 0.2 to 0.9 for subsample ratio

# Gamma (min_impurity_decrease)
gamma = list(np.arange(0, 11))  # From 0 to 10 for the gamma regularization parameter

# Fraction of features used for each tree (% from total columns of training set)
colsample_bytree = list(np.arange(1, 10)/10)  # Values from 0.1 to 0.9 for column sampling ratio

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10))  # Values between 0.001 and 10 for L1 regularization

# Hyperparameter space for XGBoost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Define XGBoost Regressor
xgb = XGBRegressor(random_state=randomstate, verbosity=0)

# Create pipeline with preprocessing and XGBoost model
estimator_xgb = Pipeline([
    ('preprocessing', transformer_reduced),  # Assuming 'transformer' is defined elsewhere for preprocessing
    ('model', xgb)
])

# Define KFold cross-validation
crossval = KFold(n_splits=5, shuffle=True, random_state=randomstate)

# Hyperparameter space (ensure it's defined as 'hyperparam_space_xgb')
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}

# Hyperparameter tuning using RandomizedSearchCV
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions=hyperparam_space_xgb,
    n_iter=50,  # Number of random combinations to try
    cv=crossval, 
    scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error', 'r2'],
    n_jobs=-1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state=randomstate
)
```

#### Gauss-Markov Assumptions Evaluation

For the regression model to meet the **Gauss-Markov assumptions** and ensure the best linear unbiased estimator (BLUE), the following conditions must hold:

1. **Normality of Residuals**:
   - **Observation**: The residuals do not follow a normal distribution.
   - **Evidence**: Both graphical analysis (Hist and QQ plot) and statistical tests (Shapiro-Wilk test) support this. The p-value obtained is significantly smaller than 0.05 (p-value = 5.02e-18), which leads to the rejection of **H0** (residuals are normally distributed). Therefore, the residuals **are not normally distributed**.

2. **Homoscedasticity**:
   - **Observation**: There is varied spread in the residuals, indicating potential heteroscedasticity.
   - **Evidence**: The scatter plot of residuals vs. fitted values reveals large variations in residuals, with some values significantly higher. This suggests that the variance of residuals is not constant, thus violating the **homoscedasticity** assumption of the Gauss-Markov theorem.

3. **Multicollinearity**:
   - **Observation**: No significant multicollinearity is detected.
   - **Evidence**: The heatmap of the correlation matrix shows that no variables have a correlation coefficient above 0.8 or below -0.8, suggesting that multicollinearity is not a problem for this model.

**Conclusion:**

- **Normality of residuals**: Not Satisfied
- **Homoscedasticity**: Not Satisfied
- **Multicollinearity**: Satisfied

Given that both **normality of residuals** and **homoscedasticity** assumptions are violated, the Gauss-Markov theorem is **not fully satisfied** in this model. Thus, while the model may still provide unbiased estimates, it may not be the most efficient, and alternative methods or adjustments may be required to improve model reliability and precision.
